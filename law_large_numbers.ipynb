```python
import numpy as np
import pandas as pd 
import os

import scipy.stats as sts 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline
```

```python
n_obs = 1000
np.random.seed(42)

x = np.random.choice(np.arange(1,7), size =n_obs)
```

```python
x_cumsum = np.cumsum(x)
x_cumsum[:10]
```

```python
x_mean = x_cumsum/np.arange(1,n_obs+1)
x_mean[:10]
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_mean)
plt.axhline(3.5, color='b', linestyle='dashed', linewidth=2)
plt.xlabel('Number of dice rolls')
plt.ylabel('Avg Value')
```

```python
n_obs = 1000000
np.random.seed(42)

x = np.random.choice(np.arange(1,7), size =n_obs)
x_cumsum = np.cumsum(x)
x_mean = x_cumsum/np.arange(1,n_obs+1)

eps = 0.001

plt.figure(figsize=(10,5))
plt.plot(x_mean)
plt.xlim(n_obs*0.9, n_obs)
plt.ylim(3.5-2*eps, 3.5+2*eps)
plt.axhline(3.5+eps, color='g', linestyle='dashed', linewidth=2)
plt.axhline(3.5-eps, color='g', linestyle='dashed', linewidth=2)

plt.xlabel('Number of dice rolls')
plt.ylabel('Avg Value')
```

```python
n_obs = 10**4

np.random.seed(42)

x = np.random.choice(np.arange(1,7), size =(n_obs,1000))
x_cumsum = np.cumsum(x, axis=0)
x_mean = x_cumsum/np.arange(1,n_obs+1)[:, None]
x_mean.shape
```

```python
eps1 = 0.1
eps2 = 0.01
mean_ = 3.5

bad_events_1 = np.abs(x_mean - mean_) > eps1
bad_events_2 = np.abs(x_mean - mean_) > eps2

proba_1 = np.mean(bad_events_1, axis=1)
proba_2 = np.mean(bad_events_2, axis=1)

plt.figure(figsize=(12,6))
plt.plot(proba_1, label = f'Prob. of range {eps1}')
plt.plot(proba_2, label = f'Prob. of range {eps2}')


plt.xlabel('Number of dice rolls')
plt.ylabel('Avg Value')
plt.legend(fontsize=14)
```

```python
cauchy_rv = sts.cauchy()

n_obs = 10**4
x = cauchy_rv.rvs(size= (n_obs,1000))
x.shape

```

```python
x_cumsum = np.cumsum(x, axis=0)
x_mean = x_cumsum/np.arange(1, n_obs+1)[:, None]
x_mean.shape
```

```python
plt.figure(figsize=(12,6))
plt.plot(x_mean[:,1])
plt.axhline(0, color='g', linestyle='dashed', linewidth=2)


plt.xlabel('Number of dice rolls')
plt.ylabel('Avg Value')
```

```python
eps1 = 0.1
eps2 = 0.01
mean_ = 0

bad_events_1 = np.abs(x_mean - mean_) > eps1
bad_events_2 = np.abs(x_mean - mean_) > eps2

proba_1 = np.mean(bad_events_1, axis=1)
proba_2 = np.mean(bad_events_2, axis=1)

plt.figure(figsize=(12,6))
plt.plot(proba_1, label = f'Prob. of range {eps1}')
plt.plot(proba_2, label = f'Prob. of range {eps2}')


plt.xlabel('Number of dice rolls')
plt.ylabel('Avg Value')
plt.legend(fontsize=14)
```
