 Ï‡Â² (chi-square test of independence)

```python
from scipy import stats
import numpy as np

observed = np.array([[30, 10],
                     [20, 40]])

chi2, p, dof, expected = stats.chi2_contingency(observed)
print(chi2, p, dof, expected)
```

ðŸ”¹ This is a very small p-value â†’ therefore, the relationship between gender and beverage choice is statistically significant.
In other words, gender influences beverage choice.

Student's t-test

```python
music = np.array([78, 82, 85, 88, 90, 76, 80])
no_music = np.array([70, 72, 68, 65, 75, 74, 73])

t_stat, p = stats.ttest_ind(music, no_music, equal_var=False)
print(t_stat, p)
```

ðŸ”¹ p < 0.05 â†’ the difference between the groups is significant.
Those who listened to music have a higher average score.

F-test (ANOVA â€” analysis of variance)

```python
method1 = np.array([78, 85, 82, 80, 77])
method2 = np.array([70, 68, 72, 69, 71])
method3 = np.array([88, 90, 85, 92, 87])

f_stat, p = stats.f_oneway(method1, method2, method3)
print(f_stat, p)
```

 Very small p â†’ the averages differ significantly between the three learning methods.
