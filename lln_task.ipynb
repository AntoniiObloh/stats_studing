```python
import numpy as np
import pandas as pd 
import os

import scipy.stats as sts 
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline
```

```python
n_obs = 10**6

norm_rv = sts.norm(loc=5, scale=3)
x = norm_rv.rvs(n_obs)
np.mean(1/x)
```

```python
n_obs = 10**6

uniform_rv = sts.uniform(0,1)
x_1 = uniform_rv.rvs(n_obs)
x_2 = uniform_rv.rvs(n_obs)
x_3 = uniform_rv.rvs(n_obs)
success = x_1 + x_2 + x_3**2 > 2
success[:5]
```

```python
np.mean(success)
```

```python
def one_beat():
    l = 1 
    m = 0
    while l > 0:
        l -=uniform_rv.rvs(1)
        m += 1
    return m
one_beat()
```

```python
n_obs = 10**4
N = [one_beat() for _ in range(n_obs)]
np.mean(N)
```
