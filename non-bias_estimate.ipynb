```python
import numpy as np
import pandas as pd 
import os

import scipy.stats as sts 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline
```

```python
norm_rv = sts.norm(loc=300,scale=100)
x_mamont = norm_rv.rvs(size=10**4)
np.mean(x_mamont)
```

```python
n = 100 
th_od = np.zeros(200)
th_pe = np.zeros(200)
th_ap = np.zeros(200)

for i in range(200):
    x_sample = np.random.choice(x_mamont, size=n, replace=False)
    x_mean = np.mean(x_sample)
    th_od[i] = x_mean
    th_pe[i] = x_mean - 4200/n
    th_ap[i] = x_mean - 5*(n+1)/n
```

```python
np.mean(th_od), np.mean(th_pe), np.mean(th_ap)
```

```python
plt.figure(figsize=(12,6))
df = pd.DataFrame({'th_od':th_od,'th_pe':th_pe,'th_ap':th_ap})
sns.boxplot(data=df);
```
