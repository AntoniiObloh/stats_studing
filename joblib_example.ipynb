```python
from joblib import Parallel, delayed
from tqdm import notebook

def simple_function(x):
    return x**10
nj = -1 #parallel for each core
result = Parallel(n_jobs=nj)(delayed(simple_function)(item)
                            for item in notebook.tqdm(range(10000000)))
```
