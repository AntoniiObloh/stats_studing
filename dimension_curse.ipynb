```python
from scipy.special import gamma

def find_pi(ndim, n_obs = 10**6):
    x = [np.random.uniform(0,1,size=n_obs) for _ in range(ndim)]

    dist = np.zeros(n_obs)
    for item in x:
        dist += item**2

    V = np.sum(dist <= 1)/n_obs

    pi = (2**ndim * V *gamma((ndim/2)+1))**(2/ndim)
    return pi
```
