```python
import numpy as np
import pandas as pd 
import os

import scipy.stats as sts 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline
```

```python
norm_rv = sts.norm(loc=300,scale=100)
x_mamont = norm_rv.rvs(size=10**6)
np.mean(x_mamont)
theta_real = np.mean(x_mamont)
theta_real
```

```python
theta_hat = np.zeros((100,50))
for n in range(100,10100,100):
    x_sample = np.random.choice(x_mamont, size = (n,50), replace=False)
    x_mean = np.mean(x_sample,axis=0)
    theta_hat[n//100 - 1] = (x_mean - 5*(n+1)/n)
    #theta_hat[n//100 - 1] = (x_mean - 4200/n)
    #theta_hat[n//100 - 1] = x_mean
theta_hat.shape
```

```python
plt.figure(figsize=(14,8))
plt.plot(theta_hat, c='grey', alpha=0.3)
plt.xlabel('Obs count',size=24)
plt.ylabel('Estimator', size=24)

plt.hlines(theta_real, 0, 100, color='blue', lw=4, label='real value')
plt.legend(fontsize=20)
```
