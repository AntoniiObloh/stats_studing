```python
import numpy as np
import pandas as pd 
import os

import scipy.stats as sts 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline
```

```python
uniform_rv = sts.uniform(0,1)

n_obs = 10**6
y = uniform_rv.rvs(n_obs)
plt.hist(y,density=True)
plt.show()
```

```python
x = -0.5*np.log(1-y)
plt.hist(x,density=True,bins=50)
plt.show()
```
